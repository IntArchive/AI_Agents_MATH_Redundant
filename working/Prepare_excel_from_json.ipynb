```python
import os
import json
import pandas as pd

# Directory containing the JSON files
json_dir = r"C:\Users\haidm18\Documents\My_git\AI_Agents_MATH_Redundant\working\Problem_WITH_RA"

# List all files in the directory matching the desired pattern
json_files = [os.path.join(json_dir, fname) for fname in os.listdir(json_dir) if fname.startswith("result_task_") and fname.endswith(".json")]

all_records = []

for filepath in json_files:
    with open(filepath, "r", encoding="utf-8") as f:
        try:
            data = json.load(f)
            # If the structure is a list of items
            if isinstance(data, list):
                all_records.extend(data)
            else:
                # Single object: wrap in a list to be consistent
                all_records.append(data)
        except Exception as e:
            print(f"Could not load {filepath}: {e}")

# Convert to dataframe
df = pd.DataFrame(all_records)

# Show a preview
print(df.head())
```

```python
df.to_excel("data.xlsx",index=False)
```
